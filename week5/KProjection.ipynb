```python
from kprojection import KProjectionCoverage
```

```python
#Odd defining
odd_description = {

    "weather": ["sunny", "rainy"],

    "lighting": ["day", "night"],

    "road_type": ["urban", "highway"]
}
```

```python
#Test case scenarios
#scenarios actually covered by our test dataset.
test_scenarios = [

    {
        "weather": "sunny",
        "lighting": "day",
        "road_type": "urban"
    },

    {
        "weather": "sunny",
        "lighting": "day",
        "road_type": "highway"
    },

    {
        "weather": "rainy",
        "lighting": "night",
        "road_type": "urban"
    }
]
```

```python
#Compute k=1 Coverage
cov1 = KProjectionCoverage(
    k=1,
    desc=odd_description
)

cov1.add_scenarios(test_scenarios)

print(cov1.compute())
```

```python
#Coverage k=2
cov2 = KProjectionCoverage(
    k=2,
    desc=odd_description
)

cov2.add_scenarios(test_scenarios)

print(cov2.compute())
```

```python
#k=3 Coverage
cov3 = KProjectionCoverage(
    k=3,
    desc=odd_description
)

cov3.add_scenarios(test_scenarios)

print(cov3.compute())
```

k=1 coverage > k=2 coverage > k=3 coverage

because:
higher combinations are harder to fully cover.
